```python
continuous_cols = ['acousticness', 'danceability', 'energy',
    'liveness', 'loudness',  'tempo',  'valence', 'speechiness', 'instrumentalness']
discrete_cols = ['key','mode','time_signature','target']
```

EXPLORATORY DATA ANALYSIS

```python
sns.set_palette("rainbow_r")
fig, axes = plt.subplots(3,3,figsize=(15,20))
axes = axes.flatten()
ax_no = 0
num_cols = continuous_cols
for col in num_cols:
    sns.histplot(data = df, x=col, bins=25, kde=True, ax = axes[ax_no])
    ax_no += 1
plt.show()
```

INSIGHTS GAINED:

PLOT 1 : ACOUSTICNESS VS COUNT :
The positive skewness in the histogram indicates an asymmetry in the relationship between the acousticness and the count . It implies that there is a tendency for the acousticness to have lower values with a smaller number of extreme or higher values.

PLOT 2: LOUDNESS VS COUNT :
The negative skewness in the histogram indicates an asymmetry in the relationship between the loudness and the count . It implies that there is a tendency for the loudness to have higher values with a larger number of extreme or higher values.

PLOT 6: TEMPO VS COUNT :
It's bimodal histogram, i.e, it has two peaks suggesting two distinct modes or clusters

PLOT 7: VALENCE VS COUNT :
It's broad curve, therefore,data points are spread out over a wide range of values. The distribution is characterized by a relatively flat or wide peak, indicating that there is no bias or concentration of values towards either end of the range. There is higher standard deviation or a larger variance in the data.

NOTE: Other Plots Provide with same insights as other plots mentioned , in terms of skewness

```python
fig, axes = plt.subplots(3,3,figsize=(15,20))
palettes = ['viridis','Set1', 'prism', 'rocket']
axes = axes.flatten()
ax_no = 0
for col in continuous_cols:
    sns.set_palette(palettes[ax_no%4])
    sns.histplot(data = df, x=col, hue='target', bins=25, kde=True, ax = axes[ax_no])
    ax_no += 1
fig.suptitle('Distributions of Continuous Features')
plt.show()
```

INSIGHTS GAINED:


1.Songs with higher acousticity are disliked and once with almost no acousticity are more likely to be liked.

2.Songs with more danceability are more likely to be liked ,that is value greater than 0.6 ,as compared to the once with lesser danceability.

3.Songs with medium energy are more likely to be liked, that is, value between 0.4 to 0.8.

4.For liveness, it can't be said for what range of values, the song would be liked or disliked.

5.Songs with lower levels of loudness are more likely to be liked, that is value below -5 (almost).

6.Songs between the range of 100 to 150 for tempo, are most likely to be liked.
7.Songs having valence greater than 0.5 are more likely to be liked.

8.Songs with speechiness between 0.1 and 0.4 are more likely to be liked.

9.Songs with instrumentalness greater than and equal to 0.1 are more likely to be liked, therefore, songs where instrumentalness is present, are more likely to be liked.



```python
sns.set_palette("Set1")
fig, axes = plt.subplots(2,2,figsize=(16,10))
palettes = ['tab10', 'Paired', 'rocket', 'Set1']
axes = axes.flatten()
ax_no = 0
for col in discrete_cols:
    sns.set_palette(palettes[ax_no%4])
    sns.countplot(data = df, x= col, ax = axes[ax_no], hue='target')
    ax_no += 1
fig.suptitle('Distributions of Discrete Features')
plt.show()
```

INSIGHTS GAINED:

1.Songs with higher key values are more likely to be disliked and songs with higer values of keys are more likely to be liked.

2.Songs having a major mode are disliked more and songs having minor mode are more liked.

3.For all values of time signature observed, the amount of liking and disliking is almost similar.

4.The number of likes and dislikes are almost equal.

```python
def danceability_segment(danceability):
  if danceability >= 0.000 and danceability < 0.400:
    return "music_less_vibrant_to_dance"
  if danceability >= 0.400 and danceability < 0.800:
    return "music_mid_vibrant_to_dance"
  if danceability >= 0.800 and danceability < 1.000:
    return "music_high_vibrant_to_dance"
  return "danceability_score_not_validated"

df['Danceability_Music_Segment'] = df['danceability'].apply(danceability_segment)
plt.figure(figsize=(25,8))
sns.countplot(x='Danceability_Music_Segment', hue ="target", data=df)
```

INSIGHTS GAINED:

1.A large amount of values of danceablility lies in the "music_mid_vibrant_to_dance" category.

2."music_high_vibrant_to_dance" is more likely to be liked as compared to being disliked.

3."music_less_vibrant_to_dance" is almost equally likely to be liked and disliked.

4."music_mid_vibrant_to_dance" is more likely to be disliked as compared to being liked.
