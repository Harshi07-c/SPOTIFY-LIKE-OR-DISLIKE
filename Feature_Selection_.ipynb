#IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

#LOADING DATA

```python
df = pd.read_csv('/content/data.csv')

print(df.shape)
display(df.head())
```

DROPPING VALUES : UNNAMED: 0 and Song Title

```python
df = df.drop(["Unnamed: 0","song_title"],axis=1)
df.head()
```

#HANDLING OF CATEGORICAL VARIABLE : "artist"

```python
import pandas as pd
import category_encoders as ce

artist = df["artist"]
target = df["target"]

# Initialize TargetEncoder
encoder = ce.TargetEncoder(artist)

# Fit and transform the data
encoded_data = encoder.fit_transform(artist, target)

print(encoded_data)
```

```python
df["artist"] = encoded_data
df.head()
```

#FEATURE SELECTION

METHOD 1 :CORRELATION BETWEEN ATTRIBUTES

```python
corr = df.corr()
corr.style.background_gradient(cmap="coolwarm")
```

METHOD 2 : INFORMATION GAIN

```python
from sklearn.feature_selection import mutual_info_classif

df1 = df.drop(['target'], axis=1)
features = df1
target = df["target"]
feature_scores = mutual_info_classif(features,target,random_state = 0)
feature_scores_sorted = sorted(zip(features.columns, feature_scores), key=lambda x: x[1])

# Print feature names and scores in ascending order
for feature, score in feature_scores_sorted:
    print(f"Feature: {feature}, Score: {score}")
```

INSIGHTS GAINED:

1.key, mode and liveness should be dropped, for their scores are 0.

2.time_signature should also be dropped as its score is very low and its importance in determining a song's quality or how "good" it will be is subjective and context-dependent. The overall appeal of a song relies on a combination of various musical elements, creativity, emotional connection, and audience preferences.



```python
df = df.drop(["time_signature" ,"key", "mode","liveness"],axis = 1)
df.head()
```
